```python
content = requests.get(URL)
soup = BeautifulSoup(content.text, 'html.parser')
row = soup.find_all('span',{"class":"bold"})
row
```

```python
x=[]
for i in row:
    x.append(i.text)
x    
```

```python
y=[]
for i in range(0,len(x)):
    if i%4 != 0:
        y.append(x[i])
y        
  
```

```python
z=[]
for i in range(1,len(y)):
    if i%3 != 0:
        z.append(y[i-1]) 
z    
        
    
            
```

```python
run=[]
for i in z:
    p=[]
    for k in i:
        if k != '/' and k !=' ':
            p.append(k)
        if k == '/' or k ==' ':
            break
    run.append(p)
run    
```

```python
runs=[]
for i in run:
    runs.append("".join(i))
runs  
```

```python
for k in range(0,len(runs)):
    if runs[k]=='':
        runs[k]=0
runs        
    
```

```python
over=[]
for k in z:
    w=[]
    for t in range(len(k)):
        if k[t]=='(':
            w.append(k[t+1])
            if k[t+2] != " ":   
                w.append(k[t+2])
            if k[t+3] != ' ' and k[t+3] != '/' and k[t+3] !='o':
                w.append(k[t+3])
                if k[t+3]=='.':
                    w.append(k[t+4])
            break
    over.append(w)        
over     
```

```python
overs=[]
for i in over:
    overs.append("".join(i))
    
overs 
```

```python
runs1=[]
runs2=[]
for i in range(len(runs)):
    if i%2 ==0:
        
        runs1.append(runs[i])
    else:
        runs2.append(runs[i])
overs1=[]
overs2=[]
for i in range(len(overs)):
    if i%2 ==0:
        
        overs1.append(overs[i])
    else:
        overs2.append(overs[i]) 
print(len(runs1))
print(len(runs2))
print(len(overs1))
print(len(overs2))




```

```python
r1=[]
for i in runs1:
    r1.append(float(i))
r2=[]
for i in runs2:
    r2.append(float(i))
```

```python
o1=[]
for i in overs1:
    if(len(i)==4):
        g=int(i[0])*10+int(i[1])+int(i[3])*0.1
        o1.append(g)
    elif (len(i)==2):
        u=int(i[0])*10+int(i[1])
        o1.append(u)
    elif (len(i)==0):
        o1.append(1)
print(len(o1)) 
print(o1)

o2=[]
for i in overs2:
    if(len(i)==4):
        g=int(i[0])*10+int(i[1])+int(i[3])*0.1
        o2.append(g)
    elif (len(i)==2):
        u=int(i[0])*10+int(i[1])
        o2.append(u)
    elif (len(i)==0):
        o2.append(1)
    elif (len(i)==3):
        j=int(i[0])+int(i[2])*0.1
        o2.append(j)
    elif (len(i)==1):
        o2.append(int(i))
print(len(o2)) 
print(o2)
```

```python
runrate1=[]
for i in range(len(o1)):
    runrate1.append((float(r1[i])/float(o1[i]))-(float(r2[i])/float(o2[i])))
runrate1                    
    
```

```python
runrate2=[]
for i in runrate1:
    runrate2.append(-i)
runrate2 
```

```python
rr2019 = pd.DataFrame(runrate1,runrate2)
rr2019.to_csv('rr2019.csv')
```
