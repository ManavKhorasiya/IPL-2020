# Model Tuning

```python
from sklearn.model_selection import train_test_split
x = data_need_2019[["Toss_Winner","Top Batsmen Points(Team A)","Top Batsmen Points(Team B)","Top Baller Points(Team A)","Top Baller Points(Team B)","Match Won after winning Toss Team A","Match Won after winning Toss Team B","runs in pp team a","runs in pp team b","Run Rate(Team A)","Run Rate(Team B)","Most Value Player Points(Team A)","Most Value Player Points(Team B)","Win ratio (Team A)","Win ratio (Team B)","Team A Ground","Team B Ground"]]
y = data_need_2019['winner']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
from sklearn.svm import SVC
svm = SVC(C = 1.7, kernel = 'rbf')
svm.fit(x_train, y_train)
```

```python
from sklearn import metrics
y_pred = svm.predict(x_test)
print("Train Accuracy:", metrics.accuracy_score(y_train, svm.predict(x_train)))
print("Test Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("F_1 Score:", metrics.f1_score(y_test, y_pred))
print("ROC AUC", metrics.roc_auc_score(y_test, y_pred))
```
