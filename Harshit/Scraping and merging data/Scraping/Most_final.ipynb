```python
import numpy as np
import pandas as pd
```

```python
point = pd.read_csv('mpp_f.csv')
```

```python
point
```

```python
y = list()
s = 0
c = 0
for i in range(13):
    for j in range(0, 12):
        if(point.at[i, str(2008 + j)] != 0):
            s += point.at[i, str(2008 + j)]
            c+=1
    y.append(s/c)
    s = 0
    c = 0
```

```python
pd.concat([point, pd.DataFrame(y, columns = ['2020'])], axis = 1).to_csv('mpp_final.csv', index = False)
```
