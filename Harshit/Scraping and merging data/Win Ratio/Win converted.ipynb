```python
import pandas as pd
```

```python
data = pd.read_csv('win_ratio.csv')
```

```python
data_1 = data.sort_values(by = ['season']).copy()
```

```python
data_1 = data_1.reset_index()
```

```python
data_1.drop(columns = ['index'], inplace = True)
```

```python
m = dict()
for i in range(2008, 2021):
    m[i] = dict()
```

```python
for i in range(966):
    m[data_1.at[i, 'season']][data_1.at[i, 'team A']][data_1.at[i, 'team B']] = data_1.at[i, 'Win_ratio_A']
    m[data_1.at[i, 'season']][data_1.at[i, 'team B']][data_1.at[i, 'team A']] = data_1.at[i, 'Win_ratio_B']
```

```python
m
```

```python
matches = pd.read_csv('Dataset_Matches.csv')
```

```python
matches
```

```python
Win_ratio_A = list()
Win_ratio_B = list()
for i in range(756):
    team1 = matches.at[i, 'team1']
    team2 = matches.at[i, 'team2']
    season = matches.at[i, 'season']
    if(team1 == 'Delhi Capitals'):
        team1 = 'Delhi Daredevils'
    if(team2 == 'Delhi Capitals'):
        team2 = 'Delhi Daredevils'
    if(team1 == 'Rising Pune Supergiant'):
        team1 = 'Rising Pune Supergiants'
    if(team2 == 'Rising Pune Supergiant'):
        team2 = 'Rising Pune Supergiants'
    Win_ratio_A.append(m[ season ][ team1 ][ team2 ])
    Win_ratio_B.append(m[ season ][ team2 ][ team1 ])
```

```python
pd.DataFrame([Win_ratio_A, Win_ratio_B]).transpose().to_csv('Win_ratio_2019.csv', index = False)
```
